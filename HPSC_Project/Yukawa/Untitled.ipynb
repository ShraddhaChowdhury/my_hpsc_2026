```python
import numpy as np
import os

k_values = [0.5, 1.0, 2.0, 5.0]

sigma_totals = []

for k in k_values:
    filename = f"./scattering_k_{k:.6f}.dat"

    if not os.path.exists(filename):
        print(f"Missing file: {filename}")
        continue

    data = np.loadtxt(filename)

    theta = data[:,0]
    ds = data[:,2]

    # integrand = ds * sin(theta)
    integrand = ds * np.sin(theta)

    # numerical integration (trapezoidal)
    sigma = 2 * np.pi * np.trapz(integrand, theta)

    sigma_totals.append((k, sigma))

    print(f"k = {k} → σ_total = {sigma:.6f}")
```

```python
import matplotlib.pyplot as plt

k_plot = [x[0] for x in sigma_totals]
sigma_plot = [x[1] for x in sigma_totals]

plt.figure(figsize=(7,5))
plt.plot(k_plot, sigma_plot, marker='o')

plt.xlabel("k")
plt.ylabel("Total cross section σ")
plt.title("Total Cross Section vs k (Yukawa)")

plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters (match your simulation)
g = 1.0
alpha = 1.0
m_values = [0.5, 1.0, 2.0, 4.0,10.0,50.0]

# r range (avoid r=0)
r = np.linspace(0.05, 3.0, 500)
plt.figure(figsize=(8,6))

for m in m_values:
    V = -(g**2) * np.exp(-alpha * m * r) / r
    plt.plot(r, V, label=f"m={m}", linewidth=2)

plt.xlabel("r")
plt.ylabel("Potential Energy V(r)")
plt.title("Yukawa Potential: PE vs r for different m")

   # 🔥 important to see decay clearly
plt.grid(True, linestyle='--', alpha=1.0)
plt.legend()

plt.tight_layout()
plt.show()
```
